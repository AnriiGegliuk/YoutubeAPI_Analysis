<h5> This function get_channel_stats() takes in two arguments, youtube and channel_id. It queries the YouTube API using the youtube object to retrieve statistics and details about a YouTube channel specified by the channel_id.
The function makes a request to the YouTube API for the specified channel_id using the channels().list() method with the specified part parameter that includes snippet, contentDetails, and statistics.
The API response is then processed to extract the desired information, such as the channel name, subscriber count, total views, total number of videos, and playlist ID for the channel uploads. 
This information is then stored in a dictionary data, which is appended to the all_data list. Finally, the all_data list is converted into a Pandas DataFrame and returned as the output of the function. </h5>

```python
def get_channel_stats(youtube, channel_id):
    
    all_data = []
    
    request = youtube.channels().list(
    part="snippet,contentDetails,statistics",
    id=''.join(channel_id)                            # Kalle Flodin Youtube channel id
    )
    response = request.execute()

    for i in response['items']:
        data = {'channelName': i['snippet']['title'],
                'subsctibers': i['statistics']['subscriberCount'],
                'views': i['statistics']['viewCount'],
                'totalVideos': i['statistics']['videoCount'],
                'playlistid': i['contentDetails']['relatedPlaylists']['uploads'],  
        }
        
        all_data.append(data)
    return(pd.DataFrame(all_data))
```

```python
play_listid = 'UUe-5wDW9r3-C0HVdcm9_knA'

def get_video_id(youtube, play_listid):
    video_id = []
    next_page_token = None
    
    while True:
        request = youtube.playlistItems().list(
            part="contentDetails",
            playlistId=play_listid,
            maxResults=50,
            pageToken=next_page_token
        )
        response = request.execute()
        
        video_id += [item['contentDetails']['videoId'] for item in response['items']]
        
        next_page_token = response.get('nextPageToken')
        if not next_page_token:
            break
    
    return video_id
```

```python
video_id = get_video_id(youtube, play_listid)
```

```python
def get_video_details(youtube, video_id):

    all_video_info = []
    
    for i in range(0, len(video_id), 50):
        request = youtube.videos().list(
            part="snippet,contentDetails,statistics",
            id=','.join(video_id[i:i+50])
        )
        response = request.execute() 

        for video in response['items']:
            stats_to_keep = {'snippet': ['channelTitle', 'title', 'description', 'tags', 'publishedAt'],
                             'statistics': ['viewCount', 'likeCount', 'commentCount', 'subscriberCount'],
                             'contentDetails': ['duration', 'definition']
                            }
            video_info = {}
            video_info['video_id'] = video['id']

            for k in stats_to_keep.keys():
                for v in stats_to_keep[k]:
                    try:
                        video_info[v] = video[k][v]
                    except:
                        video_info[v] = None

            all_video_info.append(video_info)
    
    return pd.DataFrame(all_video_info)
```

```python
# group the data by date and calculate the total view count
kf_daily_views = video_df.groupby(video_df['publishedAt'].dt.date)['viewCount'].sum()

# create a line plot with seaborn
sns.lineplot(x=kf_daily_views.index, y=kf_daily_views.values)

# set the title and axis labels
plt.title("Kalle Flodin's Daily View Count")
plt.xlabel('Year')
plt.ylabel('View Count per million views')

# display the plot
plt.show()
```

```python
# filter the DataFrame to include only rows with publishedAt between 2016 and 2019
mask = (video_df['publishedAt'] >= '2016-01-01') & (video_df['publishedAt'] <= '2019-12-31')
video_df_2016_2019 = video_df.loc[mask]

# group the data by year and calculate the total view count
kf_yearly_views = video_df_2016_2019.groupby(video_df_2016_2019['publishedAt'].dt.year)['viewCount'].sum()

# convert the index (year) to a column
kf_yearly_views = kf_yearly_views.reset_index()

# plot the total view count per year using a barplot
ax = sns.barplot(x='publishedAt', y='viewCount', data=kf_yearly_views)


# set the title and axis labels
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos:'{:,.0f}'.format(x/0.1) + ' views'))
ax.set_title("Kalle Flodin's Total View Count per Year (2016-2019)")
ax.set_xlabel('Year')
ax.set_ylabel('Total View Count (millions)')

# display the plot
plt.show()
```
